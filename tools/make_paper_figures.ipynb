## 3. Preview the figures inline (optional)

```python
from IPython.display import IFrame, display, Markdown
for name in ["Fig2_efficiency_comparison.pdf",
             "Fig3_memory_comparison.pdf",
             "Fig4_ablation_study.pdf",
             "Fig5_scalability_comparison.pdf"]:
    display(Markdown(f"### {name}"))
    display(IFrame(name, width=900, height=420))
```

## 4. Download the figures to your computer

```python
from google.colab import files
for name in ["Fig2_efficiency_comparison.pdf",
             "Fig3_memory_comparison.pdf",
             "Fig4_ablation_study.pdf",
             "Fig5_scalability_comparison.pdf"]:
    files.download(name)
```
